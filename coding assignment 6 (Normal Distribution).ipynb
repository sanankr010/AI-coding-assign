# Gaussian Distribution Plot

Here in this assignment. Ihave generated 10000 random uniform number. We set our mean as 57.5 and standard deviation as 2 so our range would be (mean - 4*s.d) to (mean + 4*s.d). Hence we are generating random between 49.5 and 65.5. On each random value when we applied gaussian pdf got the probability. Hence we plot a graph between random number their probability.   

```python
import math
import matplotlib.pyplot as plt
import random
```

```python
k=[]
for i in range(0,10000):
    n=random.uniform(49.5,65.5)
    k.append(n)
m = 57.5
sigma = 2
pie = 3.14
print(k)
```

```python
list1 = []

for i in k:
    PDFprob = (1/(2*pie)**1/2*sigma)*math.e**-((i-57.5)/((2)**1/2*sigma))**2 #Gaussian PDF
    list1.append(PDFprob)       
```

```python
plt.scatter(k,list1,0.5)
```
