# Principle of Center Limit Theorem

Principle of central Limit Theorem states that if sample size is greater than or equal to 30 the sampling Distribution became Gaussian irrespective of the distribution of Original Population.

```python
import random
import matplotlib.pyplot as plt
import math
```

```python
def fact(N,k):
    fN=1
    for i in range(1,N+1):
        fN=fN*i
    fk=1
    for i in range(1,k+1):
        fk=fk*i
    fNk=1
    for i in range(1,(N-k+1)):
        fNk=fNk*i
    return(fN/(fk*fNk))   
```

```python
list1=[]
p=0.8
N=30
l1=[]

for i in range(1000):
    
    for i in range(0,N):
        
        cdf=0
        P=random.uniform(0,1)
        x=0
        while cdf<=P:
            
            cdf+=fact(N,x)*(p**(x))*((1-p)**(N-x))
            x+=1
    l1.append(x)
print(l1)
```

```python
for i in l1:
    list1.append(i/N)
print(list1)
```

```python
list1.sort()
print(list1)
```

```python
unique=[]
for i in range(0,len(list1)-1):
    if list1[i]!=list1[i+1]:
        unique.append(list1[i])
unique.append(list1[i+1])
print(unique)
```

```python
frequency=[1]*len(unique)
c=0
for i in range(0,len(list1)-1):
    if list1[i]==list1[i+1]:
        frequency[c]+=1
    else:
        c+=1
print(frequency)       
```

```python
plt.bar(unique,frequency,0.025)
```
