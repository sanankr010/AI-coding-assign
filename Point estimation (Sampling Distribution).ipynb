```python
import random
import matplotlib.pyplot as plt
```

```python
l=[]
n=30
for i in range(1000000):
    t=[]
    for i in range(n):
        a=random.randint(1,6)
        t.append(a)
        
    l.append(t)
    #if len
        #break
#print(l)    
```

```python
prob=[]
for i in l:
    p=0
    for j in i:
        p+=j
        #print(p,end=" ")
    p=p/n
    prob.append(p)    
```

```python
def heapify(arr, n, i):
    #print(arr,n,i)
    largest = i 
    l = 2 * i + 1     
    r = 2 * i + 2     
    if l < n and arr[i] < arr[l]:
        largest = l

    if r < n and arr[largest] < arr[r]:
        largest = r

    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i] 

        heapify(arr, n, largest)

def heapSort(arr):
    n = len(arr)

    for i in range(n, -1, -1):
        heapify(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

heapSort(prob)
```

```python
unique=[]
for i in range(0,len(prob)-1):
    if prob[i]!=prob[i+1]:
        unique.append(prob[i])
unique.append(prob[i+1])
```

```python
frequency=[1]*len(unique)
c=0
for i in range(len(prob)-1):
    if prob[i]==prob[i+1]:
        frequency[c]+=1
    else:
        c+=1
        
```

```python
plt.scatter(unique,frequency,5)
```

# WITHOUT REPLACEMENT

```python
from itertools import combinations_with_replacement
n=10
perm=combinations_with_replacement([1,2,3,4,5,6],n)
l1=[]
for k in list(perm):
    l1.append(k)
print(l1)
```

```python
prob1=[]
for i in l1:
    p=0
    for j in i:
        p+=j
        #print(p,end=" ")
    p=p/n
    prob1.append(p)
print(prob1)
```

```python
prob1.sort()
```

```python
unique1=[]
for i in range(0,len(prob1)-1):
    if prob1[i]!=prob1[i+1]:
        unique1.append(prob1[i])
unique1.append(prob1[i+1])
print(unique1)
```

```python
frequency1=[1]*len(unique1)
c=0
for i in range(len(prob1)-1):
    if prob1[i]==prob1[i+1]:
        frequency1[c]+=1
    else:
        c+=1
print(frequency1)       
```

```python
plt.bar(unique1,frequency1,0.03)
```
